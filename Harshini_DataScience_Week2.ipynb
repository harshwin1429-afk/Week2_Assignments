```python
# WORKING WITH LISTS AND ARRAYS

# Creating a list.
List = [7,2,8,10,6,9,4,20,1,5,18,12]
print('The list:',List)

# Print the length of list
print('Length of the list:' ,len(List))

# Max and Min Value of the list
print('Maximum value:' ,max(List))
print('Minimum value:' ,min(List))

# Adding and removing a value from the list
List.append(16) # adds a value to the list
List.remove(18)  # removes a value from the list
print('The Updated List:', List)

# Sorting the list
asc = sorted(List) # Ascending Order of the list
desc = sorted(List,reverse = True) # Descending Order of the list

print('Ascending order:',asc)
print('Descending order:',desc)

# Converting into NumPy arrays

import numpy as np # importing numpy
arr = np.array(List) # converting list to array

# Mean,Median and SD
print('Mean:', np.mean(arr))
print('Median:',np.median(arr))
print('Standard Deviation:',np.std(arr))
```

```python
#FILE HANDLING

# Student data with name, course and marks
students = [
    ('Aswin','Data Structures', 89 ),
    ('Nayana','DAA', 73 ),
    ('Rahul', 'SQL', 69),
    ('Sabarish','SQL', 88 ),
    ('Ravishankar','Data Structures', 95 ),
    ('Tanisha','DAA', 77),
    ('Yuva','DBMS', 68)
]

# Write data to student_data.txt
with open("student_data.txt", "w") as file:
    for name, course, marks in students:
        file.write(f"{name},{course},{marks}\n")

# 3. Read and display students with marks > 75
with open("student_data.txt", "r") as file:
    print("Students with Marks Above 75:")
    for line in file:
        name, course, marks = line.strip().split(",")
        marks = int(marks)

        if marks > 75:
            print(name, course, marks)
```
